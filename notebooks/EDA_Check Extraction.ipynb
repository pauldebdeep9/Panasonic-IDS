```python
import pickle

file = open(r"D:\66 CausalMap\Panasonic-IDS\outs\20230313_MIR_rels.obj",'rb')
f_infos = pickle.load(file)
file.close()
```

```python
f_infos
```

```python
# Get Statistics
methods = ['unicausal','unicausalm','unicausal+','causenet','unicausal_any','unicausal+causenet','all']

for m in methods:
    n_evidences = 0
    n_rels = 0
    args = []
    sents = []
    for k,evidences in f_infos.items():
        found = False
        for v in evidences:
            if m=='all':
                found = True
                n_evidences+=1
                args.append(k[0])
                args.append(k[1])
                sents.append(str(v['doc_id'])+'_'+str(v['sent_id']))
            elif m=='unicausal_any':
                if ('unicausal' in v['method']) or ('unicausal+' in v['method']) or ('unicausalm' in v['method']):
                    found = True
                    n_evidences+=1
                    args.append(k[0])
                    args.append(k[1])
                    sents.append(str(v['doc_id'])+'_'+str(v['sent_id']))
            elif m=='unicausal+causenet':
                if ('unicausal' in v['method']) or ('causenet' in v['method']):
                    found = True
                    n_evidences+=1
                    args.append(k[0])
                    args.append(k[1])
                    sents.append(str(v['doc_id'])+'_'+str(v['sent_id']))
            elif m in v['method']:
                found = True
                n_evidences+=1
                args.append(k[0])
                args.append(k[1])
                sents.append(str(v['doc_id'])+'_'+str(v['sent_id']))
        if found:
            n_rels+=1
    
    print(m)
    print(f"# sents: {n_evidences}")
    print(f"# unique sents: {len(set(sents))}")
    print(f"# rels: {n_rels}")
    print(f"# nodes: {len(set(args))}")
    
```

```python
import nltk

def get_pos_tags(text):
    text = nltk.word_tokenize(text)
    return [pos for word,pos in nltk.pos_tag(text)]

get_pos_tags('scal')
```
